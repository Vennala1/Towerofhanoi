```python
def get_key(val):
    for key, value in towers.items():
        if val == value:
            return key
def tower_of_hanoi(n, source, destination, intermediate):
        destination.append(source.pop())
        print(f"Move disk 1 from {get_key(source)} to {get_key(destination)}")
        displayTowers()
        return
    tower_of_hanoi(n-1, source, intermediate, destination)
    destination.append (source.pop())
    print (f"Move disk {n} from {get_key(source)} to {get_key(destination)}")
    displayTowers()
    tower_of_hanoi(n-1, intermediate, destination, source)
def displayTowers():
    for level in range(disks,-1,-1):
        for tow in (towers['A'],towers['B'],towers['C']):
            if level>=len(tow):
                display(0)
            else:
                display(tow[level])
        print()
    #to display the names of rod
    space=" "*(disks)
    print(space+" A"+space*2+' B'+space*2+' C\n')
def display(width):
    space=" "*(disks-width)
    if width==0:
        print(space+'||'+space,end="")
    else:
        disk='@'*width
        label=str(width).rjust(2,'_')
        print(space+disk+label+disk+space,end="")
        
disks=3
A=list(range(disks,0,-1))
B=[]
C=[]
towers={'A':A,'B':B,'C':C}
displayTowers()
tower_of_hanoi(disks,A,B,C)

      
```
